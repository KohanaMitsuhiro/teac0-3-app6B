```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import yfinance as yf
from yahoo_finance_api2 import share
from yahoo_finance_api2.exceptions import YahooFinanceError

%matplotlib inline

from datetime import datetime
from datetime import timedelta
```

## toyota

```python
toyota = yf.Ticker('7203.T')
```

```python
hist = toyota.history(start = '2021-01-01',end = '2022-12-11')
hist
```

```python
hist.info()
```

## indexリセット

```python
hist = hist.reset_index()
hist
```

## Dateのデータ型をobject→datetime64へ変換

```python
hist['Date'] = pd.to_datetime(hist['Date'])
hist.info()
```

## Dateの表記変更

```python
hist.index.strftime('%Y-%m-%d')
hist
```

## floatの四捨五入

## カラムの並び替え

```python
hist = hist[['Date','High','Low','Open','Close']]
hist.head()
```

## 目的変数の追加 
- 今回予測したい翌日の終値が本日の終値よりも上がるのかどうかの情報を追加

```python
hist_shift = hist.shift(-1)
hist_shift
```

```python
#翌日の終値と本日の終値を引き算し、その結果を'delta_Close'というカラムに追記
hist['delta_Close'] = hist_shift['Close'] - hist['Close']
hist
```

```python
#目的変数Upを追加する(翌日の終値が上がる場合1、それ以外は0とする)、'delta_Close'カラムの削除
hist['Up'] = 0
hist['Up'][hist['delta_Close'] > 0] = 1
hist = hist.drop('delta_Close', axis=1)
hist
```

## カラム名の変更

```python
hist = hist.rename(columns={'Open':'Stock_Open','High':'Stock_High','Low':'Stock_Low','Close':'Stock_Close'})
hist.head()
```

## カラムの並びを戻す

```python
hist = hist[['Date','Stock_Open','Stock_High','Stock_Low','Stock_Close','Up']]
hist.head()
```

# Indexを日付に戻す

```python
hist.set_index(keys='Date', inplace=True)
hist.head()
```
